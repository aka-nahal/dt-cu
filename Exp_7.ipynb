```python
from deepface import DeepFace
import cv2 # used to read the images
import matplotlib.pyplot as plt
img1=cv2.imread("/content/caupo-768x1024.jpg")
plt.imshow(img1)
plt.show()

# BGR format :-
# bThe BGR is a 24-bit representation where the lower-addressed 8 bits are blue, the next-addressed 8 are green and higher-addressed 8 are red. RGB values tend to be written as RGB(r,g,b) where the r/g/b values are between 0 and 255 inclusive or as #rrggbb, where rr/gg/bb are 8-bit hex values.
```

```python
# plotting the image
# flipping the color from BGR to RGB
plt.imshow(img1[:, :, ::-1 ])
plt.show()
```

```python
## Analyze Image
obj=DeepFace.analyze(img1,actions = ['age','gender','race','emotion']) # analazing user-fined parameters
print(obj)
```

```python
# print result
print("Your age is",obj["age"],"and you belong to",obj["gender"],"species you are",obj["dominant_race"], "and current emotion is",obj["dominant_emotion"])
```
