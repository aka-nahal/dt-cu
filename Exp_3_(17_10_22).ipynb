```python
dataSets = get_data("index") # to get index
DiabetesDataSet = get_data("diabetes") # to get data from index
```

```python
from pycaret.classification import* # to import everything
s = setup(data=DiabetesDataSet, target='Class variable', silent=True) # to train the machine with data
rfModel = create_model('rf')  # to create and process the model from given data
plot_model(rfModel, plot='confusion_matrix') # to print/show the model genereated
```

```python
newDataSet = get_data("diabetes").iloc[:10] # to predict for a limited data
```

```python
newPredictions = predict_model(rfModel, data = newDataSet) # to take new data to the exisiting model
newPredictions
```
