```python
import matplotlib.pyplot as plt # for  graphs
x = ( 0,1,2,3,4,5,6 )
y = ( 2,3,4,7,6,7,2 )
plt.plot(x,y)
plt.title('Amplitude Vs Time') # title
plt.ylabel('Amplitude') # label
plt.xlabel('Time') # label
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [50,40,70,80,20]
y2 = [80,20,20,50,60]
plt.plot(x,y1,'blue', label="BMW", linewidth=1)
plt.plot(x,y2, 'red', label="BENZ", linewidth=1)
plt.title('Car Details In Line Plot')
plt.ylabel('Distance In Kms')
plt.xlabel('Days')
plt.legend() # to show difference and identification
plt.show()
```
